```python
import traintools
from trainconstants import *

import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Finding a lowerbound based on passengers.

df = traintools.read_schedule("datasets/nsdata1.txt")

minimum_number_of_passengers = traintools.minimum_number_of_trains_at_time_t(df, (1, 1))
print(minimum_number_of_passengers.max()[0]//sum(TYPE_3_TRAIN))

ax = minimum_number_of_passengers.plot(figsize=(20, 10))

ax.set_title('Minimum number of passengers at each point in time')
ax.set_ylabel('Passenger requirement')
ax.set_xlabel('Minutes past midnight')
ax.legend().remove()
ax.grid(True)

plt.savefig('images/naive_lowerbound.pdf', bbox_inches='tight')
```

```python
# Finding a lower bound based on minimum per trip.

minimum_number_of_passengers = traintools.minimum_number_of_trains_at_time_t(df, TYPE_3_TRAIN)
print(minimum_number_of_passengers.max()[0])

ax = minimum_number_of_passengers.plot(figsize=(20, 10))

ax.set_title('Minimum number of trains at each point in time')
ax.set_ylabel('Train requirement')
ax.set_xlabel('Minutes past midnight')
ax.legend().remove()
ax.grid(True)

plt.savefig('images/final_lowerbound.pdf', bbox_inches='tight')
```
